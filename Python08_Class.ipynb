```python
# 1번 (교안 3~5p) - 클래스 선언과 빈 클래스(pass)의 활용
class MyClass:
    x = 5

p1 = MyClass() # 설계도로부터 p1 객체 생성
print("기본 속성값 x 출력:", p1.x)

class Person:
    pass # 뼈대만 만들고 비워둘 때 에러를 막아주는 pass
```

```python
# 2번 (교안 6~8p) - __init__() 함수를 통한 초기화
class Person:
    def __init__(self, name, age):
        self.name = name  # 객체 고유의 속성 저장
        self.age = age

p1 = Person("Emil", 36)
print(f"이름: {p1.name}, 나이: {p1.age}")
```

```python
# 3번 (교안 9~12p) - self를 활용한 내부 속성 접근 및 활용
class Car:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year
        
    def display_info(self):
        # self를 붙여야 내 소유의 brand, model, year 정보를 읽을 수 있음
        print(f"차량 정보: {self.year}년식 {self.brand} {self.model}")

car1 = Car("Toyota", "Corolla", 2020)
car1.display_info()
```

```python
# 4번 (교안 18~19p) - 클래스 공유 변수와 인스턴스 독립 변수의 차이
class Person:
    species = "Human" # 모든 인스턴스가 공유하는 클래스 변수
    
    def __init__(self, name):
        self.name = name # 객체마다 독립적인 인스턴스 변수

p1 = Person("Emil")
p2 = Person("Tobias")

print(f"{p1.name}의 종: {p1.species}")
print(f"{p2.name}의 종: {p2.species}")
```

```python
# 5번 (교안 17p, 20p) - 속성 동적 추가 및 del 키워드 활용
class Person:
    def __init__(self, name):
        self.name = name

p1 = Person("Tobias")
p1.age = 25 # 설계도에 없던 age를 p1 객체에만 기습 추가!
print(f"{p1.name}의 나이: {p1.age}")

del p1.age # 추가했던 속성 제거
# print(p1.age) -> 주석을 풀고 실행하면 에러가 납니다.
```

```python
# 6번 (교안 24p) - 내부 메서드를 통한 데이터 상태 변경
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def celebrate_birthday(self):
        self.age += 1 # 나이 1살 증가 연산
        print(f"생일 축하합니다! {self.name}님은 이제 {self.age}살입니다.")

p1 = Person("Linus", 25)
p1.celebrate_birthday()
```

```python
# 7번 (교안 25p) - __str__이 있을 때와 없을 때의 차이점
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def __str__(self):
        return f"Person 객체 [이름: {self.name}, 나이: {self.age}]"

p1 = Person("Tobias", 36)
print(p1) # __str__ 덕분에 주소값 대신 텍스트가 명확하게 출력됨
```

```python
# 8번 (교안 28~29p) - 부모 클래스 선언과 자식 클래스의 상속
class Person: # 부모 클래스
    def __init__(self, fname, lname):
        self.firstname = fname
        self.lastname = lname
    def printname(self):
        print("이름:", self.firstname, self.lastname)

class Student(Person): # Person의 능력을 상속받은 자식 클래스
    pass

x = Student("Mike", "Olsen")
x.printname() # 자식 클래스엔 printname이 없지만 부모 것을 공짜로 사용함!
```

```python
# 9번 (교안 31~33p) - super() 지시어의 필수 활용법
class Student(Person): # 상단의 Person 클래스를 이어서 상속받음
    def __init__(self, fname, lname, year):
        super().__init__(fname, lname) # 부모의 생성자를 호출하여 성과 이름 세팅
        self.graduationyear = year    # 자식만의 고유 속성 추가
        
    def welcome(self):
        print(f"축하합니다! {self.firstname} {self.lastname}님은 {self.graduationyear}년도 졸업생입니다.")

s1 = Student("Mike", "Olsen", 2026)
s1.welcome()
```

```python
# 10번 (교안 36~38p) - 동일한 메서드명을 공유하는 다형성 구조
class Car:
    def move(self): print("자동차가 도로를 달립니다! 🚗")

class Boat:
    def move(self): print("보트가 바다를 항해합니다! ⛵")

class Plane:
    def move(self): print("비행기가 하늘을 날아갑니다! ✈️")

vehicle_list = [Car(), Boat(), Plane()]

for vehicle in vehicle_list:
    vehicle.move() # 문법은 같지만 결과는 다 다르게 나옴! (다형성)
```

```python
# 11번 (교안 39~41p) - __를 통한 변수 은닉 및 Getter/Setter
class Person:
    def __init__(self, name, age):
        self.name = name
        self.__age = age # 외부 접근 차단 (Private)
        
    def get_age(self): # 안전하게 읽어가게 해주는 Getter
        return self.__age
        
    def set_age(self, age): # 조건 검사 후 수정해주는 Setter
        if age > 0:
            self.__age = age
        else:
            print("에러: 나이는 음수가 될 수 없습니다!")

p1 = Person("Tobias", 25)
# print(p1.__age) -> 외부에서 직접 부르면 에러 발생!
print("Getter로 안전하게 읽은 나이:", p1.get_age())
p1.set_age(-5) # 잘못된 데이터 필터링 기능 작동
```

```python
# 12번 (교안 42~43p) - Inner Class의 선언 및 인스턴스 생성 경로
class Outer:
    def __init__(self):
        self.name = "Outer Class (외부)"
        
    class Inner:
        def __init__(self):
            self.name = "Inner Class (내부)"
        def display(self):
            print("안녕하세요, 내부 클래스의 메서드입니다.")

outer_obj = Outer()
inner_obj = outer_obj.Inner() # 외부 객체를 거쳐서 내부 객체 생성
inner_obj.display()
```
