```python
# 1번 (교안 1p, 2p) - 가장 기본적인 함수의 구조와 호출
def my_function():
    print("Hello from a function")

# 함수를 정의한 후 아래처럼 이름을 불러야 실행된다.
my_function()
my_function()  # 여러 번 호출도 가능.
```

```python
# 2번 (교안 9p, 10p) - 인수(Argument)를 전달받는 함수
def my_function(fname):  # fname은 Parameter
    print(fname + " Refsnes")

my_function("Emil")    # "Emil"은 Argument
my_function("Tobias")  # "Tobias"는 Argument
```

```python
# 3번 (교안 12p) - 매개변수에 기본값 지정하기
def my_function(country = "Norway"):
    print("I am from", country)

my_function("Sweden")
my_function()          # 값을 넣지 않으면 기본값인 "Norway"가 출력됩니다.
my_function("Brazil")
```

```python
# 4번 (교안 6p, 16p) - 계산된 값을 돌려주는 return문
def my_function(x, y):
    return x + y

result = my_function(5, 3)  # 5 + 3의 결과인 8이 result 변수에 저장됩니다.
print(result)
```

```python
# 5번 (교안 22p, 25p) - 여러 개의 숫자를 다 더해주는 *args 활용
def my_function(*numbers):
    total = 0
    for num in numbers:
        total += num
    return total

print(my_function(1, 2, 3))         # 3개짜리 계산
print(my_function(10, 20, 30, 40))  # 4개짜리 계산도 한 함수로 가능!
```

```python
# 6번 (교안 27p, 28p) - 딕셔너리 형태로 인수를 받는 **kwargs
def my_function(**myvar):
    print("All data:", myvar)
    print("Name belongs to:", myvar["name"])

my_function(name = "Tobias", age = 30, city = "Bergen")
```

```python
# 7번 (교안 31p) - 리스트 언패킹(*)을 이용한 함수 호출
def my_function(a, b, c):
    return a + b + c

numbers = [1, 2, 3]
print(my_function(*numbers))  # my_function(1, 2, 3) 선언과 동일하게 작동합니다.
```

```python
# 8번 (교안 36p) - 지역 변수(Local)와 전역 변수(Global)의 차이
x = 300  # 전역 변수

def myfunc():
    x = 200  # 지역 변수 (함수 안에서만 살아있음)
    print("Inside function:", x)

myfunc()
print("Outside function:", x)  # 바깥의 x는 여전히 300입니다.
```

```python
# 9번 (교안 38p) - global 키워드를 통한 전역 변수 수정
x = 300

def myfunc():
    global x  # 바깥에 있는 진짜 x를 쓰겠다고 선언!
    x = 200

myfunc()
print("Modified Global x:", x)  # 함수 안에서 바깥 x가 200으로 바뀌었습니다.
```

```python
# 10번 (교안 50p, 51p) - 람다(Lambda) 함수 기본 구조
# 문법: lambda 매개변수 : 수식
multiply = lambda a, b : a * b

print(multiply(5, 6))  # 결과: 30
```

```python
# 11번 (교안 56p) - 카운트다운을 수행하는 재귀 함수 구조
def countdown(n):
    if n <= 0:         # 탈출 조건
        print("Done!")
    else:
        print(n)
        countdown(n - 1)  # 자기 자신을 다시 호출

countdown(3)
```

```python
# 12번 (교안 61p, 62p) - yield를 사용하는 제너레이터(Generator)
def count_up_to(n):
    count = 1
    while count <= n:
        yield count  # 값을 하나씩 반환하고 대기 상태로 들어감
        count += 1

for num in count_up_to(3):
    print(num)
```
