```python
# 1번 (교안 1~3p) - range의 다양한 조건과 list 변환
print(list(range(10)))        # 인수가 1개면 0부터 9까지 (10 미만)
print(list(range(3, 10)))     # 인수가 2개면 시작값부터 종료값 미만
print(list(range(3, 10, 2)))  # 인수가 3개면 마지막 값은 건너뛰는 간격(step)
```

```python
# 2번 (교안 7~9p) - 리스트 요소 추가 및 삭제
cars = ["Ford", "Volvo", "BMW"]

cars.append("Honda")   # 맨 뒤에 요소 추가
cars.pop(1)            # index 1번(두 번째) 위치의 요소 삭제 ("Volvo" 제거)
cars.remove("BMW")     # 값 이름으로 직접 지우기

print("최종 리스트:", cars)
print("리스트 길이:", len(cars)) # 요소의 총 개수 출력
```

```python
# 3번 (교안 13~14p) - platform 모듈과 dir() 내장 함수
import platform as pf # platform 모듈을 pf라는 별명으로 가져옴

x = pf.system()       # 현재 운영체제(Windows, Linux 등) 확인
print("운영체제 시스템:", x)

# dir()은 모듈 안에 어떤 변수와 함수들이 들어있는지 목록으로 보여줍니다.
print("모듈 내부 멤버 일부:", dir(pf)[:5])
```

```python
# 4번 (교안 15p) - from 키워드를 사용한 임포트
from math import pi, sqrt # math 모듈에서 pi와 sqrt만 콕 집어 가져오기

print("파이 값 직접 접근:", pi)
print("제곱근 계산:", sqrt(64)) # math.sqrt()라고 안 써도 바로 작동!
```

```python
# 5번 (교안 16~17p) - 현재 시간 출력 및 특정 날짜 세팅
import datetime

now = datetime.datetime.now() # 현재 컴퓨터 시간 불러오기
print("지금 이 순간:", now)
print("올해의 연도:", now.year) # 연도만 쏙 빼오기

past_date = datetime.datetime(2020, 5, 17) # 특정 날짜 객체 만들기 (2020년 5월 17일)
print("지정된 날짜:", past_date)
```

```python
# 6번 (교안 18~19p) - strftime을 활용한 날짜 가독성 변환
import datetime

x = datetime.datetime(2018, 6, 1)

print(x.strftime("연도 코드(대문자 Y): %Y")) # 4자리 연도 (2018)
print(x.strftime("월 코드(대문자 B): %B"))   # 월 전체 이름 (June)
print(x.strftime("요일 코드(대문자 A): %A")) # 요일 전체 이름 (Friday)
```

```python
# 7번 (교안 20~21p) - 최댓값, 최솟값, 절대값, 거듭제곱
print("최솟값 찾기:", min(5, 10, 25))    # 최저치 반환
print("최댓값 찾기:", max(5, 10, 25))    # 최고치 반환
print("절대값 계산:", abs(-7.25))       # 음수를 양수로 변환
print("거듭제곱(4의 3제곱):", pow(4, 3)) # 4 * 4 * 4 연산
```

```python
# 8번 (교안 22p) - 올림과 내림 함수 비교
import math

x = math.ceil(1.4)   # 1.4보다 큰 가장 가까운 정수로 올림 -> 2
y = math.floor(1.4)  # 1.4보다 작은 가장 가까운 정수로 내림 -> 1

print("ceil(1.4) 결과:", x)
print("floor(1.4) 결과:", y)
```

```python
# 9번 (교안 24p) - JSON String -> Python Dict 전환
import json

# 따옴표 주의: 전체는 홑따옴표(''), Key와 Value는 쌍따옴표("") 구조의 JSON 문자열
json_string = '{ "name":"John", "age":30, "city":"New York"}'

# 문자열을 파이썬 변수로 파싱합니다.
py_dict = json.loads(json_string)

print("변환된 데이터 타입:", type(py_dict))
print("나이 정보 접근:", py_dict["age"])
```

```python
# 10번 (교안 25p, 29p) - Python Dict -> JSON String 가독성 옵션 전환
import json

member_dict = {
    "name": "John",
    "age": 30,
    "city": "New York"
}

# 가독성을 높이기 위해 4칸 들여쓰기(indent)를 넣고 Key 순서대로 알파벳 정렬(sort_keys)
json_result = json.dumps(member_dict, indent=4, sort_keys=True)
print(json_result)
```
