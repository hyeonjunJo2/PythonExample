```python
# 1번 (교안 2p) - re.search()를 이용한 특정 패턴 매칭
import re

txt = "The rain in Spain"

# "The"로 시작하고 중간에 아무 글자나 온 뒤 "Spain"으로 끝나는지 검사
x = re.search("^The.*Spain$", txt)

if x:
    print("패턴이 일치합니다!")
else:
    print("일치하는 패턴이 없습니다.")
```

```python
# 2번 (교안 8p) - re.findall() 사용법
import re

txt = "The rain in Spain"

# "ai"가 들어있는 모든 곳을 찾아서 리스트로 반환
x = re.findall("ai", txt)
print("찾은 결과:", x)

# 없는 단어를 찾을 때
y = re.findall("Portugal", txt)
print("없는 단어 조회 시:", y)
```

```python
# 3번 (교안 10p) - re.split()과 maxsplit 매개변수
import re

txt = "The rain in Spain"

# 공백(\s)을 기준으로 단어를 다 쪼개기
result1 = re.split("\s", txt)
print("전체 쪼개기:", result1)

# maxsplit=1을 주면 첫 번째 공백만 쪼개고 나머지는 그대로 둡니다.
result2 = re.split("\s", txt, 1)
print("한 번만 쪼개기:", result2)
```

```python
# 4번 (교안 11p) - re.sub() 패턴 치환과 count 제어
import re

txt = "The rain in Spain"

# 모든 공백(\s)을 숫자 9로 변경
result1 = re.sub("\s", "9", txt)
print("전체 치환:", result1)

# 앞에서부터 딱 2개의 공백만 숫자 9로 변경
result2 = re.sub("\s", "9", txt, 2)
print("2개만 치환:", result2)
```

```python
# 5번 (교안 17~18p) - try-except 기본형 및 특정 에러(NameError) 처리
try:
    print(x) # x를 정의하지 않았기 때문에 에러가 발생합니다!
except NameError:
    print("Variable x is not defined (변수가 정의되지 않는 에러 발생)")
except:
    print("Something else went wrong (그 외의 다른 에러 발생)")
```

```python
# 6번 (교안 20p) - f-string 기본 및 소수점 둘째 자리 제한
price = 59
txt = f"The price is {price:.2f} dollars"

print(txt)
print(f"직접 숫자 넣기: {95:.2f}")
```

```python
# 7번 (교안 21p) - f-string 내부 조건문 연산
price = 49

# price가 50보다 크면 'Expensive', 아니면 'Cheap'을 출력
txt = f"It is very {'Expensive' if price > 50 else 'Cheap'}"
print(txt)
```

```python
# 8번 (교안 23~24p) - 천 단위 구분 기호 표시
price = 59000
txt = f"The price is {price:,} dollars"
print(txt)

# 보너스: 다른 형태(예: :b는 2진수 format)
print(f"10을 2진수로 변환: {10:b}")
```

```python
# 10번 (교안 30~31p) - input() 문자열 처리를 숫자로 형변환하기
import math

# 프롬프트 메시지와 함께 입력받기
x = input("Enter a number (숫자를 입력하세요): ")

# 문자로 들어온 x를 float(실수)로 변환한 뒤 제곱근 계산
y = math.sqrt(float(x))
print(f"The square root of {x} is {y}")
```
