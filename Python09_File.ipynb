```python
# [준비 예제] 실습을 위한 기초 파일 생성하기
with open("demofile.txt", "w") as f:
    f.write("Hello! Welcome to demofile.txt\nThis file is for testing purposes.\n")
print("실습용 demofile.txt 파일 생성 완료!")
```

```python
# 1번 (교안 2~3p) - 파일 오픈 모드의 종류와 RT 기본값
# 모드를 생략하면 기본적으로 "rt" (Read, Text) 모드로 열립니다.
f = open("demofile.txt") 
print("기본 모드로 열린 파일 타입:", type(f))
f.close() # open()으로 열었으면 반드시 닫아주어야 안전합니다.
```

```python
# 2번 (교안 5p) - close()가 필요 없는 with 구문 활용
with open("demofile.txt") as f:
    data = f.read()
    print("with 구문으로 읽은 내용:\n", data)
# 블록을 빠져나오면 자동으로 파일이 close 상태가 됩니다.
```

```python
# 3번 (교안 7p) - 지정된 글자 수만큼 읽기
with open("demofile.txt") as f:
    partial_data = f.read(5) # 파일에서 딱 앞의 5글자만 읽어옴
    print("앞 5글자 출력:", partial_data)
```

```python
# 4번 (교안 8p) - readline() 및 for 반복문으로 line-by-line 읽기
print("--- 1단계: readline() 두 번 쓰기 ---")
with open("demofile.txt") as f:
    print("첫 번째 줄:", f.readline().strip())
    print("두 번째 줄:", f.readline().strip())

print("\n--- 2단계: for 루프로 한 줄씩 전부 순회하기 ---")
with open("demofile.txt") as f:
    for line in f:
        print("라인:", line.strip())
```

```python
# 5번 (교안 9p) - "a" 모드를 이용한 텍스트 덧붙이기
with open("demofile.txt", "a") as f:
    f.write("\nNow the file has more content!")

# 잘 추가되었는지 다시 읽어서 확인
with open("demofile.txt") as f:
    print(f.read())
```

```python
# 6번 (교안 10p) - "w" 모드의 덮어쓰기 특징 확인
with open("demofile.txt", "w") as f:
    f.write("Woops! I have deleted the content!")

# 결과 확인 (기존 Welcome 문장들이 전부 증발한 것을 볼 수 있습니다)
with open("demofile.txt") as f:
    print(f.read())
```

```python
# 7번 (교안 12~13p) - os 모듈을 활용한 파일 존재 여부 확인 및 삭제
import os

target_file = "demofile.txt"

if os.path.exists(target_file):
    os.remove(target_file)
    print(f"[{target_file}] 파일이 안전하게 삭제되었습니다.")
else:
    print("삭제하려는 파일이 존재하지 않습니다.")
```

```python
# 8번 (교안 11p, 15p) - "x" 모드 및 폴더 삭제 문법
import os

# 이미 위에서 demofile.txt를 지웠기 때문에 에러 없이 정상 생성됩니다.
try:
    f = open("demofile.txt", "x")
    f.write("새로 태어난 파일")
    f.close()
    print("x 모드로 새 파일 생성 성공!")
except FileExistsError:
    print("에러: 이미 똑같은 이름의 파일이 존재합니다!")

# 참고: 빈 폴더 삭제 문법 (테스트용 폴더가 없으므로 문법 형상만 확인)
# os.rmdir("myfolder")
```
