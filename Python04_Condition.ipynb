```python
# 1번 (교안 5p) - 다중 조건문(if-elif-else)을 이용한 성정 계산
# score가 75점이므로 순서대로 비교하다가 score >= 70 조건에서 "Grade: C"를 출력
score = 75

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")  # 조건 만족! 여기서 출력하고 종료
else:
    print("Grade: F")
```

```python
# 2번 (교안 14p) - 삼항 연산자(Shorthand If...Else)로 큰 값 찾기
# 문법 구조: [참일 때 값] if [조건식] else [거짓일 때 값]
a = 10
b = 20
bigger = a if a > b else b  # a가 b보다 크면 a를, 아니면 b를 bigger에 대입

print("Bigger is", bigger)  # 결과: Bigger is 20
```

```python
# 3번 (교안 17p, 22p) - 두 조건이 모두 맞아야 하는 and 연산자
# 점수가 0점과 100점 사이의 정상 범위에 있는지 검사
score = 85

if score >= 0 and score <= 100:
    print("Valid score")    # 두 조건 다 맞으므로 출력됨
else:
    print("Invalid score")
```

```python
# 4번 (교안 24p) - 중첩 조건문(Nested If) 구조
# 나이가 18세 이상인지 먼저 검사하고, 통과하면 면허가 있는지 한 번 더 검사.
age = 25
has_license = True

if age >= 18:
    if has_license:
        print("You can drive")  # 1차, 2차 조건 모두 만족 시 출력
    else:
        print("You need a license")
else:
    print("You are too young to drive")
```

```python
# 5번 (교안 28p, 29p) - 파이썬 match-case 문 활용
# day 값이 4이므로 case 4에 있는 "Thursday"를 매칭하여 출력
day = 4

match day:
    case 1: print("Monday")
    case 2: print("Tuesday")
    case 3: print("Wednesday")
    case 4: print("Thursday")
    case 5: print("Friday")
    case _ : print("Weekend") # 언더바(_)는 매칭되는 게 없을 때 쓰는 기본 fallback(else 역할)
```

```python
# 6번 (교안 30p) - while 반복문 기본 구조
# i가 6보다 작을 때까지만 반복해서 숫자를 출력
i = 1
while i < 6:
    print(i)
    i += 1  # ★중요: 이 줄이 없으면 i가 영원히 1이라 무한 루프에 빠짐
```

```python
# 7번 (교안 31p) - for 반복문을 이용해 리스트 값 하나씩 꺼내기
fruits = ["apple", "banana", "cherry"]

for x in fruits:
    print(x)  # 리스트 안의 과일 이름이 하나씩 순서대로 출력
```

```python
# 8번 (교안 32p) - break를 이용해 반복문 강제 탈출하기
fruits = ["apple", "banana", "cherry"]

for x in fruits:
    print(x)
    if x == "banana":
        break  # 과일 이름이 "banana"가 되는 순간 반복문을 바로 끝남
```
